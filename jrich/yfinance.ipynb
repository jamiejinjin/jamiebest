```python
import yfinance as yf
```

```python
stock = yf.Ticker("MSFT")
```

```python
stock.history(period="1mo")
```

```python
stock.history_metadata
```

```python
income_df = stock.income_stmt
income_df
```

```python
# remove rows with any NaN values
income_df = income_df.dropna(axis=0, how='any')
# change all 0. values to 1. to avoid dividing by 0
income_df = income_df.replace(0,1)
income_df
```

```python
income_dates = list(income_df.columns)
income_dates
```

```python
# build variation analysis for income statement

for i in range(1, len(income_dates)):
    income_df[f"change:{income_dates[i-1]}~{income_dates[i]}"] = (income_df[income_dates[i-1]] - income_df[income_dates[i]]) / income_df[income_dates[i]]
```

```python
income_df
```

```python
stock.news
```

```python
news_title=list(i['title'] for i in stock.news)
news_title
```

```python
balance_df = stock.balance_sheet
balance_df = balance_df.dropna(axis=0, how='any')
balance_df
```

```python
ticker = "MSFT"
```

```python
prompt = f"""
For the stock {ticker}:
the income statement for last 4 years is:
{income_df}
the banlance sheet for last 4 years is:
{balance_df}
the recent news for this stock is:
{news_title}
please analyze the data and make your decision
"""
```

```python
print(prompt)
```
